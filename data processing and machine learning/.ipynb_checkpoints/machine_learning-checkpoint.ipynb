```python
import numpy as np
import pandas as pd
```

```python
neurons = pd.read_csv("all_neuron_data.csv")
```

```python
# Initial transpose of data (no Cell Type column and Measurement is now the index)
new_neurons = neurons.pivot(columns="Measurement", values="Value")
new_neurons.head(10)
```

```python
current_cell = ""
current_dict = {}
big_list = []

for index, row in neurons.iterrows():
    if current_cell != row["Cell Type"]:
        # if the current cell and next cell are different, make a new empty row
        big_list.append(current_dict)
        current_dict = {}
        current_cell = row["Cell Type"]
        current_dict["Cell Type"] = current_cell
        current_dict[row["Measurement"]] = row["Value"]
    else:
        # else add the measurement value into the appropriate measurement column
        current_dict[row["Measurement"]] = row["Value"]

del big_list[0]
print(big_list)
```

```python
cleaned = pd.DataFrame(big_list)
cleaned.loc[cleaned["Cell Type"] == "Dorsal root ganglion cell"]
```

```python
# Figuring out how to rearrange columns in df
cols = list(cleaned.columns.values)
# Cell type
# [cols[6]]
# start of list up to cell type
cols[:6]
# # everything after cell type
# cols[7:]
```

```python
# Rearrange df's columns to move cell type to first, then display new df
cols = list(cleaned.columns.values)
new_cols = []
new_cols = [cols[6]] + cols[:6] + cols[7:]
cleaned = cleaned[new_cols]
cleaned
```

```python
# Number of cell types
len(cleaned["Cell Type"].unique())
```

```python
X = cleaned.drop("Cell Type", axis = 1)
y = cleaned["Cell Type"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical
from sklearn.preprocessing import Imputer

# removed stratify=y
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
# X_scaler = StandardScaler().fit(X_train)
# X_train_scaled = X_scaler.transform(X_train)
# X_test_scaled = X_scaler.transform(X_test)
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
imp = imp.fit(X_train)
X_train_imp = imp.transform(X_train)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
# error: y_test has labels (cell types) that don't exist in y_train
encoded_y_test = label_encoder.transform(y_test)

# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```
